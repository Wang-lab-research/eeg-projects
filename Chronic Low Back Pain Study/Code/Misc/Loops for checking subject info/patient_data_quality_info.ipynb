```python
sub_ID = []
num_chs = []
num_chs_drop = []
chs_drop = []
num_epos_drop = []
epos_drop = []

for file in os.listdir(data_dir):    
    if ( (not file.startswith('C')) and file.endswith("epo.fif") and \
        file[:3] in sub_fnames ):
        sub_id = file
        print(sub_id)
        
        # save
        sub_ID.append(sub_id[:3])
    
        # read data
        print(f"{sub_id}\nreading *_preprocessed-epo.fif file...")
        epochs = mne.read_epochs(data_dir+sub_id, preload=True)
    
        # save
        num_chs.append(len(epochs.info['ch_names']))
        num_chs_drop.append(len(epochs.info['bads']))
        chs_drop.append(epochs.info['bads'])
        
    if ( (not file.startswith('C')) and file.endswith("drop_log.mat") and \
        file[:3] in sub_fnames ):
        drop_log = scio.loadmat(data_dir+file)['drop_log'].tolist()
        if drop_log:
            drop_log = drop_log[0]
        
        # save
        num_epos_drop.append(len(drop_log))
        epos_drop.append(drop_log)
    
    else:
        continue
        
    display.clear_output(wait=True)
```

```python
# dictionary of lists 
dict = {'sub_ID': sub_ID, '# Channels Total': num_chs, 
        '# Channels Dropped': num_chs_drop, 'Channels Dropped': chs_drop, 
        '# Epochs Dropped': num_epos_drop, 'Epochs Dropped': epos_drop } 
    
df = pd.DataFrame(dict)
    
df.head()
```

```python
# save as excel file
excel_fname = "patient_data_quality_info.xlsx"
df.to_excel( os.path.join(save_dir,excel_fname) )
```
