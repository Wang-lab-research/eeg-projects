```python
# subject ID
data_path = "../../Data/Processed Data/"

sub_id=''
unequal_subs_lst=[]

for sub_id in sub_ids_lst:
    sub_data_path = os.path.join(data_path,sub_id)

    try: 
        # Process only subjects with 64 channel data, for homogeneity
        print(f"Subject {sub_id}\n")

        # print(f"{sub_num}\nreading preprocessed-epo file...")
        epochs = mne.read_epochs(sub_data_path+'_preprocessed-epo.fif', preload=False)
        display.clear_output(True)

        # print(f"\n*len(epochs) =\t\t{len(epochs)}*\n")

        # Load in sample times for stimuli
        epo_times = scio.loadmat(sub_data_path+'_epo_times.mat')
        epo_times_raw = epo_times['epo_times'] # leave as array
        epo_times = [el[0] for el in epo_times_raw]
        # print(f"\n*len(epo_times) =\t{len(epo_times)}*\n")

        stim_labels = scio.loadmat(sub_data_path+'_stim_labels.mat')
        stim_labels = stim_labels['stim_labels'].tolist()
        if type(stim_labels[0]) == list and len(stim_labels[0])==1:
            stim_labels = [el[0] for el in stim_labels]
        else:
            stim_labels = stim_labels[0]
        # print(f"\n*len(stim_labels) =\t{len(stim_labels)}*")

        # Load in pain rating for each stimuli
        pain_ratings = scio.loadmat(sub_data_path+'_pain_ratings.mat')
        pain_ratings = pain_ratings['pain_ratings'].tolist()[0]
        # print(f"*len(pain_ratings) =\t{len(pain_ratings)}*\n")

        # Load in drop log for bad trials
        drop_log = scio.loadmat(sub_data_path+'_drop_log.mat')
        drop_log = drop_log['drop_log'] # leave as array
        if drop_log.shape[0] > 0:
            drop_log = drop_log[0]
        # print(f"*len(drop_log) =\t{drop_log.shape[0]}*\n")

        # print(len(epochs))
        # print(len(stim_labels))
        
        if len(epochs) == len(stim_labels):
            print("All equal")
        else:
            print("Mismatch !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
            unequal_subs_lst.append(sub_id)

    except:
        continue
        
```
